```python
import cv2
import os
import random
import matplotlib.pylab as plt
from glob import glob
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
PATH = os.path.abspath(os.path.join('.'))
FOLDERS_PATH = os.path.join(PATH, "train")
SOURCE_IMAGES = []
images = []

for src in os.listdir(FOLDERS_PATH):
    folder_path = os.path.join(FOLDERS_PATH, src)
    if os.path.isdir(folder_path):
        SOURCE_IMAGES.append(folder_path)
        images = images + glob(os.path.join(folder_path, "*.jpg"))

images[:11]
```

```python
random_images = random.sample(images, 3)

print('Path: {0}'.format(random_images[0]))
plt.figure(figsize=(16,16))
plt.subplot(131)
plt.imshow(cv2.imread(random_images[0]))

print('Path: {0}'.format(random_images[1]))
plt.subplot(132)
plt.imshow(cv2.imread(random_images[1]))

print('Path: {0}'.format(random_images[2]))
plt.subplot(133)
plt.imshow(cv2.imread(random_images[2]));
```
